```python
import pandas as pd 

#in UBUNTU:
Scores=pd.read_csv("/media/vahid/New Volume/Python_learning/Pandas_Learning/Scores.csv"
                    ,index_col=0)
#in WINDOWS: 
#Scores=pd.read_csv("E:\Python_learning\Pandas_Learning\Scores.csv",
#                   index_col=0)
print(Scores,'\n\n')


# Native accessors:
# there are two way of indexing in dictionaries:
Fizik_Scores=Scores.Fizik     #by doting
Arabi_Scores=Scores['Arabi']  #by []
print(Fizik_Scores,'\n\n',   
      Arabi_Scores,'\n\n')       
#for accesing special value of data we do: 
Amir_Fizik_gread=Scores.Fizik['Amir']
Ghasem_Arabi_gread=Scores['Arabi']['Ghasem']
print(" Amir's Fizik greade: ",Amir_Fizik_gread,'\n',   

      "Ghasem's Arabi gread: ",Ghasem_Arabi_gread)


### Indexing in pandas:
# there are two ways of indexing in Pandas: 


## First way: Index-based selection: 
#To select the specific row of data in a DataFrame based on number of row and column:  
Ehsan_Score=Scores.iloc[4]
print("Ehsan's Scores: ",'\n',Ehsan_Score,'\n\n')
# to get a specific column with iloc, we can do the following:
Zists_Scores=Scores.iloc[:,5]
print("Zists' Scores: ",'\n',Zists_Scores,'\n')
three_first_zist_Scores=Scores.iloc[:3,5]
print("three first Zists' Scores: ",'\n',three_first_zist_Scores,'\n')
#to access special datas: 
Special_Scores=Scores.iloc[[0,2,5],1]
print("special Scores: ",'\n',Special_Scores,'\n')
#to access the four last Scores: 
four_last_Scores=Scores.iloc[-4:]
print("four last Scores: ",'\n',four_last_Scores,'\n')
Mahmood_Riazi_grade=Scores.iloc[3,0]
print("Mahmood Riazi grade: ",Mahmood_Riazi_grade,'\n')

## second way: Label-based selection:
#To select the specific row of data in a DataFrame based on Label of row and column:  
Amir_Scores=Scores.loc['Amir']
print("Amir's Scores: ",'\n',Amir_Scores,'\n')
Saeid_Zaban_grade=Scores.loc['Saeid','Zaban']
print("Saeid's Zaban Scores: ",Saeid_Zaban_grade,'\n')

##Consideration:
#  iloc uses the Python stdlib indexing scheme, where the first element
#  of the range is included and the last one excluded.
#  So 0:10 will select entries 0,...,9.
#  loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

## set_index method can change index by replacing column we want to be index. 
print('setting Riazi as index:','\n',Scores.set_index("Riazi"),'\n')

## Conditional Selection: 
# we can select part of data conditionally, that is, choosing part of data 
#based on a condition. like what came aftewards: 
Who_passed_Riazi=Scores.Riazi>=10
print('Who Passed Riazi: ','\n',Who_passed_Riazi,'\n')
#This operation produced a Series of True/False booleans 
# This result can then be used inside of loc to select the relevant data:
Who_passed_Riazi2=Scores.loc[Scores.Riazi>=10]
print('Scores of persons Who Passed Riazi: ','\n',Who_passed_Riazi2,'\n')
# We can use the ampersand (&) to bring the two questions together:
Who_passed_Riazi_and_Fizik=Scores.loc[(Scores.Riazi>=10) & (Scores.Fizik>=10)]
print('Scores of persons Who Passed Riazi and Fizik: ','\n',Who_passed_Riazi_and_Fizik,'\n')
#sometimes we want to use or For this we use a pipe (|):
Who_passed_Arabi_or_Zaban=Scores.loc[(Scores.Arabi>=10) | (Scores.Zaban>=10)]
print('Scores of persons Who Passed Arabi or Zaban: ','\n',Who_passed_Arabi_or_Zaban,'\n')
# we have other selector too:
#  isin is lets you select data whose value "is in" a list of values.
Who_passed_Riazi_by_12_15=Scores.loc[Scores.Riazi.isin([12,15])]
print('Scores of persons Who Passed Riazi by 12 and 15: ','\n',Who_passed_Riazi_by_12_15,'\n')
#The second is isnull (and its companion notnull).
# These methods let you highlight values which are (or are not) empty (NaN).
who_attending_Zaban_Exame=Scores.loc[Scores.Zaban.notnull()]
print('Scores of persons who attending Zaban Exame: ','\n',who_attending_Zaban_Exame,'\n')


## Assigning data:
# You can assign either a constant value:
Scores_tempt=Scores
Scores_tempt['Arabi']=0
print(Scores_tempt)
# Or with an iterable of values:
Scores_tempt['Fizik']=range(20, 13, -1)
print(Scores_tempt)
```

```python
import pandas as pd

Scores=pd.read_csv("/media/vahid/New Volume/Python_learning/Pandas_Learning/Scores.csv"
                    ,index_col=0)


## Summary functions: 
# .describe() method: This method generates a high-level summary of the attributes of the given column.
describe_Scores=Scores.describe()
print(describe_Scores,'\n')
# for describing specific coloumn: 
describe_Riazi_Scores=Scores.Riazi.describe()
print(describe_Riazi_Scores,'\n')
# to see the mean of coloumn datas: 
mean_Zaban_scores=Scores.Zaban.mean()
print('Mean of Zaban scores: ',mean_Zaban_scores,'\n')
#To see a list of unique values we can use the unique() function:
Values_of_Zist=Scores.Zist.unique()
print('Values of Zist','\n',Values_of_Zist,'\n')
# by .value_counts we can have a counts of values: 
Values_of_Zist_counter=Scores.Zist.value_counts()
print('Values_of_Zist_counter','\n',Values_of_Zist_counter,'\n')


## Maps: 
# There are two mapping methods that you will use often.
# map(): for instance we want to remean the scores
# of Zist to 0: 
Scores_Zist_mean=Scores.Zist.mean()
print("mean of Zist's Scores: ", Scores_Zist_mean,'\n')
maped_Zist_Scores=Scores.Zist.map(lambda p: p-Scores_Zist_mean)
print(maped_Zist_Scores,'\n')


############### not completed: kaggle Leson3: 
# apply() is the equivalent method if we want to transform a whole
#  DataFrame by calling a custom method on each row.
Scores_Zaban_mean=Scores.Zaban.mean()
def remean_Scores(row):
    row.Zist=row.Zist-Scores_Zaban_mean
    return row

Scores.apply(remean_Scores)
```
